```python
class CNN2(nn.Module):
    def __init__(self,input_dim,num_classes,device):
        super(CNN2, self).__init__()
        # kernel
        self.input_dim = input_dim
        self.num_classes = num_classes

        conv_layers = []
        conv_layers.append(nn.Conv1d(in_channels=1,out_channels=64,kernel_size=3,padding=1)) # ;input_dim,64
        conv_layers.append(nn.BatchNorm1d(64))
        conv_layers.append(nn.ReLU(True))

        conv_layers.append(nn.Conv1d(in_channels=64,out_channels=128,kernel_size=3,padding=1)) #(input_dim,128)
        conv_layers.append(nn.BatchNorm1d(128))
        conv_layers.append(nn.ReLU(True))

        self.conv = nn.Sequential(*conv_layers).to(device)

        fc_layers = []
        fc_layers.append(nn.Linear(input_dim*128,num_classes))
        self.classifier = nn.Sequential(*fc_layers).to(device)

    def forward(self, x):
        batch_size, D = x.shape
        x = x.view(batch_size,1,D)

        x = self.conv(x)
        x = torch.flatten(x,1)
        x = self.classifier(x)
        return x
```

```python
def load_checkpoint(filepath):
        if os.path.isfile(filepath):
            checkpoint = torch.load(filepath)
            print("Model loaded with batch_size = {}, seen {} epochs and {} mini batches".
                format(checkpoint['batch_size'],checkpoint['epoch'],checkpoint['batch'])) 
            return checkpoint
        else:
            return None
        
            
def load_model(model,filepath,inference_mode=True):
        #filepath = join(self.runs_dir,'checkpoint.pth')
        checkpoint = load_checkpoint(filepath)
        
        model.load_state_dict(checkpoint['state_dict'])
        
        if inference_mode:
            for parameter in model.parameters():
                parameter.requires_grad = False
            model.eval()
        return model
    
def predict(model,x,device,eval_mode=False):
        tensor_x = torch.stack([torch.Tensor(i) for i in x]).to(device)
        
        pred = torch.zeros(0,dtype=torch.int64).to(device)
        
        model.eval()        
        
        with torch.no_grad():
                outputs = model(tensor_x)
                _, predi = torch.max(outputs.data,1)
                pred = torch.cat((pred,predi))

        return pred.cpu().numpy()
```

```python

#hyper-params
batch_size = 5*1024 # increasing batch size with more gpu added
optim = 'Adam'

input_dim = 76
num_class = len(np.unique(label_names))
device = torch.device('cpu:0')

model = CNN2(input_dim=input_dim,num_classes=num_class,device=device)
filepath = '1D-CNN_conv_2_fc_1_4th_run_optim_Adam_lr_0.001_reg_1e-06_bs_5120.pth'
model = load_model(model,filepath)

pred = predict(model,demo_X,device,eval_mode=True)
for i in range(pred.shape[0]):
    print('True Class:{:20} --> Predicted Class: {:20} '.format(label_names[demo_y[i]],label_names[pred[i]]))
```
